```python
# Jika diketahui n adalah 2 digit terakhir dari NIM kalian, maka tentukan bilangan prima antara 1000n dan 1000(n+1)
# n = 23
def adalah_prima(n):
    if n < 2:
        return False
    for i in range(2, int(n ** 0.5) + 1):
        if n % i == 0:
            return False
    return True

start = 1000 * 23
end = 1000 * (23 + 1)

n = [i for i in range(start, end) if adalah_prima(i)]

print("Bilangan prima antara", start, "dan", end, "adalah:")
print(n)
```
